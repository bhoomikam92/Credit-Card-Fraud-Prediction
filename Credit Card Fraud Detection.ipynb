# Undersampling

Build a sample dataset from original dataset, containing similar distribution of legit and fraud transactions

Number of Fraud transactions : 492


```python
legit_sample = legit_transactions.sample(n=492)
```

Concatenate both samples

```python
new_dataset = pd.concat([legit_sample,fraud_transactions],axis = 0) #axis = 0, fraud transactions will be added below , axis =1 = columns next to each other
```

# Splitting Data into Features and Targets

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

# Splitting Data into Training and Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2, stratify = Y, random_state= 2)
#80% data will be stores in X_train and corresponding (0,1) will be stored in Y_Train same for test rest 20% of the data
#Stratify means equally distributed 0,1
#random_state reprouduce a code as it is
```

### Training the Logistic Regression using the training data

```python

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipe = make_pipeline(
    StandardScaler(with_mean=True, with_std=True),
    LogisticRegression(
        solver="lbfgs",          # good default
        max_iter=1000,           # higher than default 100
        n_jobs=-1,               # parallel where applicable
        random_state=42
    )
)
pipe.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = pipe.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = pipe.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
